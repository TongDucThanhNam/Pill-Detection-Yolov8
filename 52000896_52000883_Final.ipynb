```python
import os

# current_path
print(os.getcwd())
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO('best.pt')  # load a pretrained model (recommended for training)

# Train the model
```

```python
data_predict_path = "data_predict"
files = os.listdir(data_predict_path)
print(f'files: {files}')
```

```python
results_list = []
for file in files:
    results = model(
        os.path.join(data_predict_path, file),
        save=True
    )  # predict on an image
    results_list.append(results)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
for results in results_list:
    for result in results:
        print("result")
        # plot result
        img = result.orig_img  # ndarray
        names = result.names  # dict
        boxes = result.boxes  # Tensor | ndarray
    
        # Create figure and axes
        fig, ax = plt.subplots(1)
    
        # Create a Rectangle patch
        # print(f"boxes: {boxes}")
        data_boxes = boxes.data
        # print(f"data_boxes: {data_boxes}")
        data_boxes_numpy = data_boxes.cpu().numpy()
        # print(f"data_boxes_numpy: {data_boxes_numpy}")
        for box in data_boxes_numpy:
            print(f"box: {box}")
            x = box[0]
            y = box[1]
            w = box[2] - box[0]
            h = box[3] - box[1]
            # print(f"x: {x}, y: {y}, w: {w}, h: {h}")
            rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
            # Add the patch to the Axes
            ax.add_patch(rect)
    
        # Display the image (color)
        ax.imshow(img)
        plt.show()
```
